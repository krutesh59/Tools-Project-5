```python
#Load Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
# Load the dataset with a different encoding
youtube_data = pd.read_csv('youtube_dataset.csv', encoding='latin-1')

country_counts = youtube_data['country'].value_counts().head(3)

plt.figure(figsize=(6, 6))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 3 Countries with Highest Number of YouTube Users')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
channel_views = youtube_data.groupby('name')['videoviews'].sum().nlargest(10)
plt.figure(figsize=(8, 6))  # Set the figure size
channel_views.plot(kind='bar', color='skyblue')  # Create a bar plot
plt.title('Top 10 Channels with Highest Number of Watched Videos')  # Set the title of the plot
plt.xlabel('Channel')  # Set the label for the x-axis
plt.ylabel('Total Video Views')  # Set the label for the y-axis
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()  # Show the plot
```

```python
import matplotlib.pyplot as plt

# Grouping the data by channel type and calculating the mean video views
channel_views = youtube_data.groupby('channeltype')['videoviews'].mean()

# Plotting the histogram
plt.figure(figsize=(12, 6))
plt.bar(channel_views.index, channel_views.values, color='skyblue')
plt.xlabel('Channel Type')
plt.ylabel('Mean Video Views')
plt.title('Mean Video Views by Channel Type')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select the top 10 most subscribed channels
top_10_subscribed = youtube_data.sort_values('subscribers', ascending=False).head(10)

# Create a scatter plot of number of videos uploaded vs number of views
plt.figure(figsize=(12, 6))
plt.scatter(top_10_subscribed['uploads'], top_10_subscribed['videoviews'])
plt.xlabel('Number of Videos Uploaded')
plt.ylabel('Number of Views on Videos')
plt.title('Relationship between Number of Videos Uploaded and Number of Views on Videos for Top 10 Most Subscribed Channels')
plt.grid(True)
plt.show()
```
